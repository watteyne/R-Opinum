# Revue détaillée des timezones dans Datahub
## Environnement de test
### Compte
Nous utilisons le compte 959, Calculated Variables
### Sites
Nous nous basons sur 3 sites
* UTC, qui est en ... UTC
* Local, qui est en CE(S)T
* FarAway, qui est en HKT (Hong Kong, pas de changement d'heure)
### Sources
Nous nous basons sur x sources par site (certaines sont redondantes)
* UTC, qui est en ... UTC
* Site_timezone qui a la même timezone que son site
* Brussels, qui est en CE(S)T
* Site_timezone_adapted, qui est en UTC lors du chargement de données puis passe en la timezone du site
* UTC_shared, qui est en UTC et est partagée par les 3 sites
* Brussels_shared, qui est en CE(S)T et est partagée par les 3 sites
# Variables
Sur chaque source, nous avons 5 variables
* unaware, qui reçoit les données sans mention de timezone
* unaware_brussels, qui reçoit les données, préalablement mises en CE(S)T,  sans mention de timezone
* utc, qui reçoit les données en UTC en format Zulu (Z)
* utc_2, qui reçoit les données en UTC en format +00:00
* brussels, qui reçoit les données en CE(S)T
